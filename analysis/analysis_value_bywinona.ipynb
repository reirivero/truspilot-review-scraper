```python
# Función para analizar el sentimiento
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

# Aplicar la función a los comentarios
df['sentiment'] = df['paragraph'].fillna('', inplace=False).apply(get_sentiment)

# Resumen de los resultados
df['sentiment'].describe()
```

```python
from collections import Counter
from nltk.corpus import stopwords
# import string

# Descargar stopwords
nltk.download('stopwords')
nltk.download('punkt_tab')

# Función para limpiar y tokenizar texto
def tokenize(text):
    tokens = nltk.word_tokenize(text)
    tokens = [word.lower() for word in tokens if word.isalpha()]
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return tokens

# Aplicar la función a los comentarios
df['tokens'] = df['paragraph'].fillna('', inplace=False).apply(tokenize)

# Contar la frecuencia de palabras
all_tokens = [token for tokens in df['tokens'] for token in tokens]
word_freq = Counter(all_tokens)

# Mostrar las palabras más comunes
print(word_freq.most_common(10))
```

# 2. Correlación entre Calificaciones y Comentarios
Para analizar la correlación entre las calificaciones y las palabras o frases utilizadas en los comentarios, puedes usar scikit-learn para vectorizar los textos y calcular la correlación:

```python
from sklearn.feature_extraction.text import CountVectorizer
from scipy.stats import pearsonr


# Vectorizar los comentarios
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['paragraph'].fillna('', inplace=False))

# Calcular la correlación entre cada palabra y la calificación
correlations = {}
for i, word in enumerate(vectorizer.get_feature_names_out()):
    correlations[word] = pearsonr(X[:, i].toarray().flatten(), df['rated'])[0]

# Mostrar las palabras con mayor correlación
sorted_correlations = sorted(correlations.items(), key=lambda x: x[1], reverse=True)
print(sorted_correlations[:10])
```
